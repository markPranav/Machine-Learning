## AIM: Polynomial Regression on Air Quality to predict Absolute Humidity

### Description

Polynomial Regression is a regression algorithm that models the relationship between a dependent(y) and independent variable(x) as nth degree polynomial. The Polynomial Regression equation is given below:

`y= b0+b1x1+ b2x12+ b2x13+...... bnx1n`

It is also called the special case of Multiple Linear Regression in ML. Because we add some polynomial terms to the Multiple Linear regression equation to convert it into Polynomial Regression.

It is a linear model with some modification in order to increase the accuracy.
The dataset used in Polynomial regression for training is of non-linear nature.
It makes use of a linear regression model to fit the complicated and non-linear functions and datasets.

Hence, "In Polynomial regression, the original features are converted into Polynomial features of required degree (2,3,..,n) and then modeled using a linear model."

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn
from matplotlib import style
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PolynomialFeatures 
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import operator
```

```python
data_set= pd.read_csv('AirQualityUCI.csv', sep=";" , parse_dates= ['Date','Time'])
data_set.describe()
```

```python
data_set.head()
```

## Data Preprocessing
### **Dropping NaN values and columns**

```python
data_set.dropna(how="all",axis=1,inplace=True)
data_set.dropna(how="all",axis=0,inplace=True)
data_set.shape
```

```python
data_set.tail()
```

```python
cols = list(data_set.columns[2:])
cols
```

Since decimal seperator by default is "," instead of "." we need to modify the float values.

```python
for col in cols:
    if data_set[col].dtype != 'float64':
        str_x = pd.Series(data_set[col]).str.replace(',','.')
        float_X = []
        for value in str_x.values:
            fv = float(value)
            float_X.append(fv)

            data_set[col] = pd.DataFrame(float_X)

data_set.head()
```

## Feature Selection

Here Date and Time are irrelevant to our prediction model

```python
features=list(data_set.columns)
```

```python
corr = data_set.corr()
corr.style.background_gradient(cmap='coolwarm')
```

```python
features.remove('Date')
features.remove('Time')
features.remove('AH')
features
```

```python
X = data_set[features]
Y = data_set['AH']
```

```python
X.head()
```

### **Train Test Split**

```python
X_train, X_test, y_train, y_test = train_test_split(X,Y, test_size=0.4, random_state=0)
```

## Linear Regression Model

```python
model = LinearRegression()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
```

### Error and Score

```python
mse = mean_squared_error(y_test,y_pred)
rmse = mean_squared_error(y_test,y_pred,squared=False)
r2 = r2_score(y_test,y_pred)
print("MSE: ",mse)
print("RMSE: ",rmse)
print("r2: ",r2)
```

**Comaprision of predicted results with actaul results**

```python
df_preds = pd.DataFrame({'Actual': y_test.squeeze(), 'Predicted': y_pred.squeeze(), 'Loss': abs(y_test.squeeze()-y_pred.squeeze())})
df_preds
```

## Polynomial Regression Model

```python
polynomial_features= PolynomialFeatures(degree=2)
x_poly = polynomial_features.fit_transform(X_train)
model = LinearRegression()
model.fit(x_poly, y_train)
```

```python
x_pred_poly = polynomial_features.fit_transform(X_test)
y_poly_pred = model.predict(x_pred_poly)
```

### Error and Score

```python
pr_train,pr_test = model.score(x_poly , y_train), model.score(x_pred_poly , y_test)
print(f"Training Score: {pr_train}")
print(f"Test Score: {pr_test}")
```

```python
mse = mean_squared_error(y_test,y_poly_pred)
rmse = mean_squared_error(y_test,y_poly_pred,squared=False)
r2 = r2_score(y_test,y_poly_pred)
print("MSE: ",mse)
print("RMSE: ",rmse)
print("r2: ",r2)
```

**Comaprision of predicted results with actaul results**

```python
df_preds = pd.DataFrame({'Actual': y_test.squeeze(), 'Predicted': y_poly_pred.squeeze(), 'Loss': abs(y_test.squeeze()-y_poly_pred.squeeze())})
df_preds
```
